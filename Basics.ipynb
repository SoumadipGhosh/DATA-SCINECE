# Concat datasets 

```python
df1=pd.DataFrame({
    "name":["SG","SS","SK"],
    "marks":[90,34,89],
})    
```

```python
df1
```

```python
df2=pd.DataFrame({
    "name":["PM","SM","TM"],
    "marks":[85,89,91],
}) 
```

```python
df2
```

```python
pd.concat([df1,df2])
```

# Merge

```python
df1=pd.DataFrame({
    "name":["SG","SS","SK"],
    "marks":[90,34,89],
})
```

```python
df2=pd.DataFrame({
    "name":["SG","SS","SK"],
    "roll":[1,2,3],
})
```

```python
pd.merge(df1,df2,on="name")
```
