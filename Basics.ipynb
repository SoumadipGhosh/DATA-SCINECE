# AXIS 

#AXIS 0 -- > COLUMN

#AXIS 1 -- > ROW

```python
x = [[1,2 ,3],[4,5,6],[7,8,9]]
```

```python
arr=np.array(x)
```

```python
arr
```

```python
arr.sum(axis=0) # sum of each column
```

```python
arr.sum(axis = 1) # sum of each row
```

```python
arr.T # transpose of array .. # rows become columns and columns become rows
```

```python
arr.flat
```

```python
arr.size
```
